```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Housing.csv")
df.head()
```

```python
print(df.shape)
df.info()
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    print("\n", col)
    print(df[col].value_counts())
```

```python
#
eda = df.copy()
mapping={'yes':1,'no':0}
for c in ['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']:
    eda[c]=eda[c].map(mapping)
eda['furnishingstatus']=eda['furnishingstatus'].map({
    'unfurnished':0,
    'semi-furnished':1,
    'furnished':2
})
print(eda.corr(numeric_only=True))
```

```python
num_cols=['price','area','bedrooms','bathrooms','stories','parking']
for col in num_cols:
    plt.figure(figsize=(5,2.5))
    plt.boxplot(df[col], vert=False)
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
clean = df.copy()

mapping={'yes':1,'no':0}
for c in ['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']:
    clean[c]=clean[c].map(mapping)

clean.rename(columns={'guestroom':'guest_room'}, inplace=True)

clean['furnishingstatus']=clean['furnishingstatus'].map({
    'unfurnished':0,
    'semi-furnished':1,
    'furnished':2
})

clean['logprice']=np.log(clean['price'])
clean.drop(columns=['price'], inplace=True)

clean.head()
```

```python
clean.to_csv("Housing_cleaned.csv", index=False)
```
